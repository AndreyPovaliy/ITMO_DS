## **Matplotlib — это одна из самых популярных библиотек для визуализации данных в Python. Она позволяет строить графики различной сложности, от простых линий до сложных трехмерных визуализаций.**

Основные особенности Matplotlib:

	•	Позволяет создавать статические, анимированные и интерактивные графики.
	•	Гибкость в настройке внешнего вида графиков.
	•	Интеграция с различными библиотеками Python (NumPy, Pandas, SciPy).
	•	Поддержка различных форматов вывода (PNG, PDF, SVG и др.).
	•	Возможность работы напрямую в Jupyter Notebook и других интерактивных средах.

**2. Основные модули**

Matplotlib состоит из нескольких ключевых модулей, наиболее важные из них:

	•	pyplot — основной модуль для построения графиков.
	•	figure и axes — отвечают за создание фигур (canvas) и отдельных осей.
	•	cm — работа с цветами и colormap.

## Пример простого графика

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Генерируем данные
x = np.array([1, 2, 3, 4, 5])
y = x + 2
y1 = x ** 2

# Создаем график
plt.plot(x, y, label="линейная ф-ция", color="green", marker="*", linestyle="--")
plt.plot(x, y1, label="квадр ф-ция", color="purple", marker="o", linestyle="-.")

# Добавляем заголовок графика и подписи осей
plt.title("Линейный график")
plt.xlabel("X")
plt.ylabel("Y")

# Включаем легенду
plt.legend()

# Включаем сетку
plt.grid(True, linestyle="--", alpha=0.2)

# Отображаем график
plt.show()
```

# Столбчатая диаграмма

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 8]

plt.bar(categories, values, color='purple')
plt.show()
```

**Настройки столбчатой диаграммы**

```python
# Данные
categories = ["Январь", "Февраль", "Март", "Апрель", "Май"]  # Категории (метки оси X)
values = [10, 15, 7, 12, 9]  # Значения для каждой категории

# Стиль графика
plt.style.use("ggplot")  # Используем стиль оформления "ggplot"

# Построение столбчатой диаграммы
plt.bar(categories, values,
        color=["blue", "red", "green", "purple", "orange"],  # Устанавливаем цвета для каждого столбца
        edgecolor="black",  # Цвет границы столбцов
        linewidth=1.5,  # Толщина границы
        alpha=0.8)  # Прозрачность столбцов (1 - непрозрачный, 0 - полностью прозрачный)

# Подписи заголовка и осей
plt.title("Продажи по месяцам", fontsize=14, fontweight="bold", color="darkblue")  # Заголовок графика
plt.xlabel("Месяцы", fontsize=12, color="purple")  # Подпись оси X
plt.ylabel("Количество продаж", fontsize=12, color="purple")  # Подпись оси Y

# Добавление значений на вершины столбцов
for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha="center", fontsize=12, fontweight="bold")
# Проходим по каждому столбцу и добавляем текст с числом чуть выше столбца

#plt.text(2,10,"Hello!", fontsize=30 )

# Настройки осей
plt.ylim(0, max(values) + 5)  # Устанавливаем диапазон оси Y (чуть выше максимального значения)
plt.yticks(np.arange(0, max(values) + 5, 5))  # Устанавливаем отметки на оси Y с шагом 5

# Добавление сетки
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.7, color="gray")
# Включаем сетку только по оси Y (чтобы не пересекала столбцы)

# Сохранение графика в файл
plt.savefig("bar_chart.png", dpi=300, bbox_inches="tight")
# Сохраняем график в файл "bar_chart.png" с высоким качеством

# Отображение графика
plt.show()  # Выводим график на экран
```

# Гистограмма

```python
data = np.random.randn(1000)
plt.hist(data, bins=100, alpha=1)
plt.show()
```

**Настройки гистограммы**

```python

# Генерация случайных данных (имитация распределения)
np.random.seed(42)  # Фиксируем случайность для воспроизводимости
data = np.random.randn(1000) * 10 + 50  # 1000 значений, нормальное распределение со средним 50 и отклонением 10

# Стиль графика
plt.style.use("ggplot")  # Используем стиль оформления "ggplot"

# Построение гистограммы
plt.hist(data, bins=20,  # Разбиваем данные на 20 интервалов (корзин)
         color="blue", edgecolor="black", linewidth=1.2, alpha=0.7)
# Задаем цвет столбцов, черную границу, толщину линии и прозрачность

# Подписи заголовка и осей
plt.title("Распределение данных", fontsize=14, fontweight="bold", color="darkblue")  # Заголовок графика
plt.xlabel("Значения", fontsize=12, color="purple")  # Подпись оси X
plt.ylabel("Частота", fontsize=12, color="purple")  # Подпись оси Y

# Добавление сетки
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.7, color="gray")
# Включаем сетку только по оси Y, чтобы не мешала восприятию

# Сохранение графика в файл
plt.savefig("histogram.png", dpi=300, bbox_inches="tight")
# Сохраняем гистограмму в файл "histogram.png" с высоким качеством

# Отображение графика
plt.show()  # Выводим график на экран
```

## Диаграмма(график) рассеивания

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', alpha=0.5)
plt.show()
```

**Настройка диаграммы рассеивания**

```python
# Генерация случайных данных
np.random.seed(42)  # Фиксируем случайность для воспроизводимости
x = np.random.rand(100) * 10  # 100 случайных точек для оси X (от 0 до 10)
y = 2.5 * x + np.random.randn(100) * 2  # Линейная зависимость с небольшим шумом

# Стиль графика
plt.style.use("ggplot")  # Используем стиль оформления "ggplot"

# Построение диаграммы рассеивания
plt.scatter(x, y,
            color="blue", edgecolor="black", linewidth=1,  # Цвет точек, черная обводка
            alpha=0.5, s=200, marker="o", label="Данные")  # Прозрачность, размер точек, форма маркера и подпись

# Подписи заголовка и осей
plt.title("Диаграмма рассеивания", fontsize=14, fontweight="bold", color="darkblue")  # Заголовок графика
plt.xlabel("Переменная X", fontsize=12, color="purple")  # Подпись оси X
plt.ylabel("Переменная Y", fontsize=12, color="purple")  # Подпись оси Y

# Добавление сетки
plt.grid(True, linestyle="--", linewidth=0.5, alpha=0.7, color="gray")  # Включаем сетку с пунктирными линиями

# Добавление легенды
plt.legend(loc="upper left", fontsize=10, frameon=True, shadow=True)  # Легенда в левом верхнем углу

# Сохранение графика в файл
plt.savefig("scatter_plot.png", dpi=300, bbox_inches="tight")  # Сохраняем график в файл с высоким качеством

# Отображение графика
plt.show()  # Выводим график на экран
```

# 5. Круговая диаграмма

```python
labels = ['Python', 'C++', 'Java', 'JavaScript']
sizes = [25, 25, 25, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'yellow'])
plt.show()
```

**Настройка круговой диаграммы**

```python

# Данные
categories = ["Apple", "Samsung", "Xiaomi", "Huawei", "Other"]  # Названия категорий
values = [35, 30, 15, 10, 10]  # Доли рынка (в процентах)

# Цвета для сегментов
colors = ["red", "blue", "green", "purple", "orange"]

# Стиль графика
plt.style.use("ggplot")  # Используем стиль оформления "ggplot"

# Построение круговой диаграммы
plt.pie(values, labels=categories,  # Данные и подписи категорий
        colors=colors,  # Задаем цвета сегментов
        autopct="%1.1f%%",  # Отображаем процент в каждом сегменте (с 1 знаком после запятой)
        startangle=140,  # Угол поворота диаграммы для лучшей читаемости
        wedgeprops={"edgecolor": "black", "linewidth": 1},  # Граница сегментов (черная, 1px)
        pctdistance=0.85,  # Расстояние подписи процентов от центра (0 - в центре, 1 - на границе)
        textprops={"fontsize": 12, "color": "black"})  # Настройка шрифта подписей

# Добавление центрального круга для создания эффекта "пончика"
centre_circle = plt.Circle((0, 0), 0.70, fc="white")  # Создаем белый круг в центре
plt.gca().add_patch(centre_circle)  # Добавляем круг на график

# Заголовок
plt.title("Рыночная доля брендов смартфонов", fontsize=14, fontweight="bold", color="darkblue")

# Делаем диаграмму круговой (чтобы не было искажений)
plt.axis("equal")

# Сохранение графика в файл
plt.savefig("pie_chart.png", dpi=300, bbox_inches="tight")  # Сохраняем график в файл с высоким качеством

# Отображение графика
plt.show()  # Выводим график на экран
```

## Работа с несколькими графиками

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

x = np.linspace(0, 10, 100)
y1, y2, y3, y4 = np.sin(x), np.cos(x), np.tan(x), np.exp(x)

axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title("Синус")

axes[0, 1].plot(x, y2, color='red')
axes[0, 1].set_title("Косинус")

axes[1, 0].plot(x, y3, color='green')
axes[1, 0].set_title("Тангенс")

axes[1, 1].plot(x, y4, color='purple')
axes[1, 1].set_title("Экспонента")

plt.tight_layout()  # Автоматическая подгонка
plt.show()
```

## Ящик с усами

```python
# Создание данных
df = pd.DataFrame({
    "Category": np.random.choice(["A", "B", "C"], 100),
    "Values": np.random.randn(100) * 10 + 50
})

# Построение графика
sns.boxplot(x="Category", y="Values", data=df)

plt.title("Ящик с усами")
plt.show()
```

📌 Основные элементы Boxplot

Вот разбор всех элементов боксплота:


1️⃣ Ящик (Box)

	•	Верхняя граница = 75-й процентиль (Q3)
	•	Нижняя граница = 25-й процентиль (Q1)
	•	Ширина ящика показывает межквартильный размах (IQR): IQR = Q3 - Q1

2️⃣ Полоса внутри ящика (Median)

	•	Это медиана (Q2, 50-й процентиль) — центральное значение данных.

3️⃣ “Усы” (Whiskers)

	•	Простираются до 1.5 * IQR от краев ящика.
	•	Если данных мало, усы могут доходить до максимального/минимального значения (но без выбросов).

4️⃣ Выбросы (Outliers)

	•	Значения, выходящие за пределы 1.5 * IQR от ящика.
	•	Обозначаются точками за пределами усов.
