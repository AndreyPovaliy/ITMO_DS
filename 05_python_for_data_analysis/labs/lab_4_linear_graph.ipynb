**Задание 1: График функции
y
=
2
x
+
3**

Требуется построить график линейной функции y=2x+3 на интервале от 5 до
-5
.

```python
# символьное вычисление
x= symbols('x')
f = 2*x + 1
plot(f)

# вычисление на отрезке
x = np.arange(-5,6)
y = 2*x + 1
plt.plot(x, y, label="линейная ф-ция", color="green", marker="*", linestyle="--")
plt.show()
```

**Задание 2: Сравнение двух функций**

Необходимо построить графики двух линейных функций
y=x и y=2x+1 на одном графике на интервале от
−10 до 10

Обе линии должны иметь разные цвета и стили линий.

```python
# символьное вычисление
x= symbols('x')
f = 2*x + 1
f1 = x
plot(f,f1)

# вычисление на отрезке
x = np.arange(-10,11)
y = 2*x + 1
y1=x
plt.plot(x, y, label="линейная ф-ция", color="green", marker="*", linestyle="--")
plt.plot(x, y1, label="линейная ф-ция", color="purple", marker="o", linestyle="-.")
plt.show()
```

**Задание №3. Сравнение четырех функций**

Необходимо построить графики четырех функций:


y=x^2
y=4x+2
y=-x
y=x^4+x^2-1


4 графика на одной фигуре на интервале от
0 до 10. Должен получиться квадрат.

Линии должны иметь разные цвета и стили линий, а также каждый сабграфик должен иметь свою легенду.

```python
# ваш код здесь
x = np.arange(11)

fig, axes = plt.subplots(2, 2, figsize=(8, 8))


y1= -x**2
y2= 4*x + 2
y3= -x
y4= x**4+x**2-1

axes[0, 0].plot(x, y1, color='blue')
axes[0, 1].plot(x, y2, color='red')
axes[1, 0].plot(x, y3, color='green')
axes[1, 1].plot(x, y4, color='purple')
plt.tight_layout()
plt.show()
```
