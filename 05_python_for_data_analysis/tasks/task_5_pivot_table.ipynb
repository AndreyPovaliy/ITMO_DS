# **–ó–∞–¥–∞–Ω–∏–µ: –∏—Å–ø–æ–ª—å–∑—É—è —Ç–æ–ª—å–∫–æ —Å–≤–æ–¥–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã `pivot_table` —Ä–µ—à–∏—Ç—å –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –∑–∞–¥–∞—á–∏**

```python
# 1. –ü–æ–ª –∏ –ø–æ–∫—É–ø–∫–∏: —Å—Ä–µ–¥–Ω–∏–π –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —á–µ–∫ –≤ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏–≥—Ä –∏—Å—Ö–æ–¥—è –∏–∑ –ø–æ–ª–∞ –∫–ª–∏–µ–Ω—Ç–∞.
pd.pivot_table(df, values="good_price", index="gender_cd", columns="category_name", aggfunc=["mean", "max"])

```

```python
# 2. –ú–µ–¥–∏–∞–Ω–Ω—ã–π –∏ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –ø–æ –≥–æ—Ä–æ–¥—É –∏ —É—Ä–æ–≤–Ω—é –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
df['amount'] = df['good_price'] * df['good_cnt']
pd.pivot_table(df, values="amount", index="city_nm", columns="education_level", aggfunc=["median", "mean"])
```

```python
# 3. –°—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç, –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç, –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–∞–∂–¥–æ–π –∏–≥—Ä–æ–≤–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏

pd.pivot_table(df, values="age", index="gender_cd", columns="category_name", aggfunc=["median", "mean",'max'])
```

```python
# 4. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∏–≥—Ä (–¢–û–ü - 10 –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø—Ä–æ–¥–∞–∂)
pd.pivot_table(df, values="good_cnt", index = "good_name",  aggfunc= "count").sort_values(by= 'good_cnt', ascending= False).head(10)
```

```python
# 5. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∏–≥—Ä –≤ –∂–∞–Ω—Ä–∞—Ö –≠–∫—à–µ–Ω, –®—É—Ç–µ—Ä—ã, –°–∏–º—É–ª—è—Ç–æ—Ä—ã (–¢–û–ü - 20 –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø—Ä–æ–¥–∞–∂)
df1= df[df['category_name'].isin(['–≠–∫—à–Ω', '–®—É—Ç–µ—Ä—ã', '–°–∏–º—É–ª—è—Ç–æ—Ä—ã'])]
pd.pivot_table(df1, values="good_cnt", index = "good_name",  aggfunc= "count").sort_values(by= 'good_cnt', ascending= False).head(20)


```

```python
# 6. –†–µ–π—Ç–∏–Ω–≥ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø—Ä–æ–¥–∞–∂ —Å—Ä–µ–¥–∏ –¥–µ—Ç–µ–π –∏ –≤–∑—Ä–æ—Å–ª—ã—Ö (–¢–û–ü - 5 –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø—Ä–æ–¥–∞–∂)
df['ageGroup'] = ['child' if x < 18 else 'adult' for x in df['age']]
pd.pivot_table(df, values="good_cnt", index = "good_name", columns= "ageGroup",  aggfunc= "count").sort_values(by="adult", ascending= False).head(5)
```

```python
pd.pivot_table(df, values="good_cnt", index = "good_name", columns= "ageGroup",  aggfunc= "count").sort_values(by="child", ascending= False).head(5)
```

```python
df['order_day'] = pd.to_datetime(df['order_day'])
```

```python
# 7. –°—Ä–µ–¥–Ω–∏–π —á–µ–∫ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –∏–≥—Ä –≤ –º–∞–µ 2022 –≥–æ–¥–∞ –∏ –º–∞–µ 2023 –≥–æ–¥–∞.
df1 = df[(df.order_day == '2022-05') | (df.order_day == '2023-05')]
pd.pivot_table(df1, values="amount", index = "good_name",  aggfunc= "mean")

```

```python
# 8 –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∏–≥—Ä –≤ –∂–∞–Ω—Ä–µ —Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ –∑–∞ 2023 –≥–æ–¥ —Å—Ä–µ–¥–∏ –º—É–∂—á–∏–Ω –∏ –∂–µ–Ω—â–∏–Ω —Ä–∞–∑–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
df1 = df[(df.order_day == '2023') | (df.category_name == '–°–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ')]
pd.pivot_table(df1, values="good_cnt", index = "gender_cd", columns= "education_level",  aggfunc= "count")

```

```python
# 9. –¢–æ–ø-10 –∫–ª–∏–µ–Ω—Ç–æ–≤, –ø–æ—Ç—Ä–∞—Ç–∏–≤—à–∏—Ö –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –¥–µ–Ω–µ–≥ –Ω–∞ –ø–æ–∫—É–ø–∫—É –∏–≥—Ä –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ü—Ä–µ–¥–∑–∞–∫–∞–∑
df1 = df[df.category_name == '–ü—Ä–µ–¥–∑–∞–∫–∞–∑—ã']
pd.pivot_table(df1, values="amount", index = "good_name",  aggfunc= "count").sort_values(by="amount", ascending= False).head(10)
```
