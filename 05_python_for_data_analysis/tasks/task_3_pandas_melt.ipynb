## Задание №3 Решение структруных проблем датасетов


✅ 1) Решите структурную проблему в данном датафрейме,

✅ 2) Удалите данные, которые не несут полезной информации.

Даны 6 складов.

В колонках в качестве переменных указано, что может хранить данный склад,

В значения - площади для хранения конкретных товаров в квадратных метрах.

**`empty_now`** - является ли склад сейчас пустым.

```python
# данную ячейку нужно запустить, код в ней менять нельзя

import numpy as np
import pandas as pd


df = pd.DataFrame(
    {
        'warehouse': ['A', 'B', 'C', 'D', 'E','F'],
        'industrial_chemistry': [20000,0,0,50000, 0, np.nan],
        'food': [0,25000,50000,0, 0, np.nan],
        'pharmacy': [0,5000,0,0, 0, np.nan],
        'home_chemistry': [10000,0 ,0 ,10000, 0, np.nan],
        'auto_supply': [0,0,0,0, 25000, np.nan],
        'empty_now': ['yes','no','no','no', 'yes', np.nan]
    }
)

df
```

```python
# 1) Решение структурной проблемы в данном датафрейме
id_vars = ["warehouse","empty_now"]
formatted_df = pd.melt(df, id_vars= id_vars, var_name="type", value_name="square")
formatted_df
```

```python
# 2) Удаление данных, которые не несут полезной информации
# formatted_df.drop(formatted_df['square'].isna())

formatted_df.drop(formatted_df[formatted_df['empty_now'].isna()].index, inplace=True)
formatted_df.drop(formatted_df[formatted_df['square'] == 0].index, inplace=True)
formatted_df
```
