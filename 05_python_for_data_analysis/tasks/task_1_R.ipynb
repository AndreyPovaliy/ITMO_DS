# Усечённое среднее

Усечённое среднее вычисляется путем отбрасывания $ p $ наименьших и $ p $ наибольших значений из упорядоченной выборки. Пусть даны отсортированные значения:

$
x_{(1)}, x_{(2)}, \dots, x_{(n)}
$

где $ x_{(1)} $ — наименьшее значение, а $x_{(n)} $ — наибольшее.

Тогда усечённое среднее с параметром $ p $ определяется как:

$
\bar{x}_c = \frac{1}{n - 2p} \sum_{i=p+1}^{n-p} x_{(i)}
$

где:

- $\ n $ — общее число элементов в выборке,  
- $\ p $ — количество отбрасываемых значений с каждого края,  
- $x_{(i)} $ — упорядоченные значения.

Этот метод позволяет снизить влияние выбросов на расчет среднего значения.



```
Например, в международных состязаниях по прыжкам в воду

является среднеарифметический балл трех оставшихся судей.
Такой подход не дает одному судье манипулировать баллом,
```
