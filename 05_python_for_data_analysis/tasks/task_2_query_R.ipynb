# **Признаки датасета "Титаник" с их описанием:**

**PassengerId:** Уникальный идентификатор пассажира.

**Survived:** Выжил ли пассажир (1 - да, 0 - нет).

**Pclass:** Класс пассажира (1 - Первый класс, 2 - Второй класс, 3 - Третий класс).

**Name:** Имя пассажира.

**Sex:** Пол пассажира (male - мужской, female - женский).

**Age:** Возраст пассажира.

**SibSp:** Количество братьев/сестер/супругов на борту.

**Parch:** Количество родителей/детей на борту.

**Ticket:** Номер билета.

**Fare:** Стоимость билета.

**Cabin:** Номер каюты.

**Embarked:** Порт отправления (C - Cherbourg, Q - Queenstown, S - Southampton).

**Задание №2**

1) Выберите всех пассажиров первого класса: Найдите всех пассажиров, которые ехали в первом классе.

2) Найдите всех выживших пассажиров: Отберите строки, где пассажиры выжили .

3) Фильтрация по возрасту: Выберите всех пассажиров младше 18 лет.

4) Отбор по порту отправления: Найдите всех пассажиров, которые сели на корабль в Саутгемптоне.

5) Поиск по фамилии: Найдите всех пассажиров с фамилией, начинающейся на "Smith".

6) Отбор по стоимости билета: Отберите всех пассажиров, стоимость билета которых была больше 50.

7) Фильтрация по полу: Найдите всех пассажиров женского пола.

8) Отбор пассажиров с несколькими членами семьи на борту: Выберите всех пассажиров, у которых было больше 2 членов семьи на борту.

9) Поиск пассажиров с номером каюты: Найдите всех пассажиров, у которых указан номер каюты (Cabin не пустое значение).

```r
# 1) Выберите всех пассажиров первого класса: Найдите всех пассажиров, которые ехали в первом классе.
subset(titanic,titanic$Pclass == 1)
```

```r
# 2) Найдите всех выживших пассажиров: Отберите строки, где пассажиры выжили .
subset(titanic,titanic$Survived == 1)
```

```r
# 3) Фильтрация по возрасту: Выберите всех пассажиров младше 18 лет.
subset(titanic, titanic$Age < 18)
```

```r
# 4) Отбор по порту отправления: Найдите всех пассажиров, которые сели на корабль в Саутгемптоне.
subset(titanic, titanic$Embarked == 'S')
```

```r
# 5) Поиск по фамилии: Найдите всех пассажиров с фамилией, начинающейся на "Smith".
subset(titanic, grepl('Smith', titanic$Name))

```

```r
# 6) Отбор по стоимости билета: Отберите всех пассажиров, стоимость билета которых была больше 50.
subset(titanic, titanic$Fare > 50)
```

```r
# 7) Фильтрация по полу: Найдите всех пассажиров женского пола.
subset(titanic, titanic$Sex == 'female')
```

```r
# 8) Отбор пассажиров с несколькими членами семьи на борту: Выберите всех пассажиров, у которых было больше 2 членов семьи на борту.
titanic$families <- titanic$SibSp + titanic$Parch
subset(titanic, titanic$families > 2)
```

```r
# 9) Поиск пассажиров с номером каюты: Найдите всех пассажиров, у которых указан номер каюты (Cabin не пустое значение).
subset(titanic, !is.na(titanic$Cabin))
```

**Задание № 3**

1) Выберите всех выживших пассажиров первого класса: Найдите всех пассажиров, которые ехали в первом классе и выжили.

2) Найдите всех несовершеннолетних, путешествовавших с родителями: Выберите пассажиров младше 18 лет, у которых было хотя бы одно количество родителей на борту.

3) Пассажиры с дорогими билетами и большим возрастом: Найдите всех пассажиров, возраст которых больше 50 лет и стоимость билета которых превышает 100.

4) Женщины, которые выжили и ехали из Квинстауна: Выберите женщин, которые выжили и сели на корабль в Квинстауне.

5) Мужчины третьего класса, путешествовавшие с братьями и сестрами: Найдите всех мужчин, которые ехали в третьем классе и имели хотя бы одного брата или сестру на борту.

6) Пассажиры второго класса с отсутствующим возрастом: Найдите всех пассажиров второго класса, у которых возраст не указан (Age is null).

7) Мужчины старше 30 лет, которые выжили: Выберите всех мужчин старше 30 лет, которые выжили.

8) Дети, ехавшие с обоими родителями: Найдите всех детей младше 12 лет, у которых было как минимум два родителя на борту.

9) Пассажиры старше 60 лет, ехавшие в первом классе, которые не выжили: Выберите всех пассажиров старше 60 лет, которые ехали в первом классе и не выжили.

```r
# 1) Выберите всех выживших пассажиров первого класса: Найдите всех пассажиров, которые ехали в первом классе и выжили.
subset(titanic, titanic$Survived == 1 & titanic$Pclass == 1)
```

```r
# 2) Найдите всех несовершеннолетних, путешествовавших с родителями: 
# Выберите пассажиров младше 18 лет, у которых было хотя бы одно количество родителей на борту.
subset(titanic, titanic$Age < 18 & titanic$Parch != 0)
```

```r
# 3) Пассажиры с дорогими билетами и большим возрастом: 
# Найдите всех пассажиров, возраст которых больше 50 лет и стоимость билета которых превышает 100.
subset(titanic, titanic$Age > 50 & titanic$Fare > 100)
```

```r
# 4) Женщины, которые выжили и ехали из Квинстауна: Выберите женщин, которые выжили и сели на корабль в Квинстауне.
subset(titanic, titanic$Sex == 'female' & titanic$Embarked == 'C')
```

```r
# 5) Мужчины третьего класса, путешествовавшие с братьями и сестрами: 
# Найдите всех мужчин, которые ехали в третьем классе и имели хотя бы одного брата или сестру на борту.
subset(titanic, titanic$Sex == 'male' & titanic$Pclass == 3 & SibSp != 0)
```

```r
# 6) Пассажиры второго класса с отсутствующим возрастом: Найдите всех пассажиров второго класса, у которых возраст не указан (Age is null).
subset(titanic, is.na(titanic$Age) & titanic$Pclass == 2)
```

```r
# 7) Мужчины старше 30 лет, которые выжили: Выберите всех мужчин старше 30 лет, которые выжили.
subset(titanic, titanic$Sex > 30 & titanic$Survived == 1)
```

```r
# 8) Дети, ехавшие с обоими родителями: Найдите всех детей младше 12 лет, 
# у которых было как минимум два родителя на борту.
subset(titanic, titanic$Age <12 & titanic$Parch == 2)
```

```r
# 9) Пассажиры старше 60 лет, ехавшие в первом классе, которые не выжили: 
# Выберите всех пассажиров старше 60 лет, которые ехали в первом классе и не выжили.
subset(titanic, titanic$Age > 60 & titanic$Pclass == 1 & titanic$Survived == 0)
```
